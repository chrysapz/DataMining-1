Here we define functions that we will use throughout our implementation.

```python
#Finds the unique values from a column
def unique_vals(df, col):
  unique_values = df[col].unique()
  return unique_values 

#This function is used to convert attributes with boolean type from string to bool
#All missing ("unknown") values will become NaN
def to_boolean(x):
 return True if x.lower() in ['yes', 'mu', '1', 'ja'] else False if x.lower() in ['no', 'sigma', '0', 'nee'] else None

#Returns max number of consecutive Nones in an attribute
def max_consecutive_nones(column):
    count = max_count = 0
    #print(column.value)
    for value in column:
        if value is None:
            count += 1
            max_count = max(max_count, count)
        else:
            count = 0
    return max_count

#Returns the total amount of invalid entries
def total_invalid(col):
  count=0
  for value in col:
    if value is None:
      count +=1 
  return count

# Print the percentage of missing values per column
def percent_missing(missing_values):
  missing_percentage = (missing_values / len(df)) * 100
  print("\nPercentage of Missing Values in Column:")
  print(missing_percentage)

#Displays the data types contained in each column.
def column_val_types(df):
  for col in df.columns:
    unique_values = unique_vals(df, col)
    unique_types = {type(value) for value in unique_values}
    print(f"Column {col} value types: {unique_types}")

#Remaps the numbers of columns so that they are still sequential after the deletion of a colum
def remap(df):
  col_map = {name: num for num, name in enumerate(df.columns)}
  df = df.rename(columns=col_map)
  return df
```

The following cell removes incorrect values from the birthday attribute. The function pd.to_datetime finds the missing values (the values that could not be transformed into datetime format). Additionally, we consider as incorrect the values that lie beyond the extreme values we have set. Both incorrect and missing values are replaced with the average birthday.

```python
birthday_col = 8

# Convert date strings into datetime, when incorrect entries are encountered they are turned into 'NaT'
df.iloc[:, birthday_col] = pd.to_datetime(df.iloc[:, birthday_col], errors='coerce')

min_date = pd.Timestamp('1957-01-01')
max_date = pd.Timestamp('2004-01-01')
valid_dates = df.iloc[:, birthday_col].dropna()

# Calculate the interquartile range (IQR) for the valid dates
q1 = valid_dates.quantile(0.25)
q3 = valid_dates.quantile(0.75)
iqr = q3 - q1

# Define the lower and upper bounds for the date range
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

def random_date_in_range(start_date, end_date):
    delta = end_date - start_date
    random_days = random.randrange(delta.days)
    return start_date + pd.Timedelta(days=random_days)

count_bday = 0

def birthday_attr(date):
    global count_bday
    if pd.isna(date) or date < min_date or date > max_date:
        count_bday += 1
        return random_date_in_range(lower_bound, upper_bound).date()
    else:
        return date.date()

# Apply the custom function to the birthday column
df.iloc[:, birthday_col] = df.iloc[:, birthday_col].apply(birthday_attr)

print("The column about the students' birthday has a total of invalid values: ",count_bday )
percent_missing(count_bday)

print(df.iloc[304, 8])
```

```python
student_count_col = 8

def student_count_attr(x):
    x = str(x)
    x = re.sub(r'[^0-9a-zA-Z\s]', '', x).strip()
    if x.isdigit() or x.replace('.', '', 1).isdigit():
        new_x = int(x)
        if new_x<0:
            return None
        else:
            return new_x

    try: 
        num = w2n.word_to_num(x)
        return int(num)
    except ValueError:
        pass
    return None

df.iloc[:, student_count_col] = df.iloc[:, student_count_col].apply(student_count_attr)

invalids_class = total_invalid(df[8])
print("The column about the students' estimation of people in class has a total of invalid values: ", invalids_class)
percent_missing(invalids_class)


max_value_sc = 500
min_value_sc = 100

valid_values = df[student_count_col][df[student_count_col].notna()]
avg_value = valid_values.mean()  # calculate the mean of non-null values

# replace null values with the mean of non-null values
#df.loc[df[student_count_col].isnull(), student_count_col] = avg_value

count_count =0
count_atlim = 0
def student_count_cleaning(x):
    global count_count
    global count_atlim
    if pd.isnull(x):
        count_count+=1
        return avg_value
    elif x < min_value_sc:
        count_count+=1
        count_atlim +=1
        return min_value_sc
    elif x > max_value_sc:
        count_atlim +=1
        count_count+=1
        return max_value_sc
    else:
        return x

df.iloc[:, student_count_col] = df.iloc[:, student_count_col].apply(student_count_cleaning)

print("The column about the students' stress level has a total of invalid values: ", count_count, count_atlim)
percent_missing(count_count)
percent_missing(count_atlim)
```
